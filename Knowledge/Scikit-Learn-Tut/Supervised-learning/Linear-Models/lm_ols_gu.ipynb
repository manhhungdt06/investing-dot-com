#### Handle GBPUSD data

```python
def handle_file(filepath: str = None) -> pd.DataFrame:
    colnames = ['Time', 'open', 'high', 'low', 'close', 'volume']
    cwd = os.getcwd()
    df = pd.read_csv(os.path.join(cwd, filepath),
                     names=colnames, sep="\\t", engine="python",
                     skipfooter=1, index_col=0)
    df.index = pd.to_datetime(df.index)
    # df['hour'] = df.index.hour
    # df['minute'] = df.index.minute
    # df.index = df.index.date
    return df
```

```python
def return_df(
        df: pd.DataFrame = None,
        len: int = 5,
        cum: bool = True,
) -> pd.DataFrame:
    import pandas_ta as ta

    # with the day before
    df['log_ret'] = ta.log_return(df['close'])*100
    df['log_ret_5'] = ta.log_return(df['close'], length=len)*100          # with a 5 day shift

    # df['log_ret_20'] = ta.log_return(df['close'], length=20)*100          # with a 20 day shift, for test
    # df['a_log5cum'] = ta.above(df['log_ret_5'], df['log_ret_20'])

    df['log_retcum'] = ta.log_return(df['close'], cumulative=cum)*100  # with a pivot
    df['a_log5'] = ta.above(df['log_ret'], df['log_ret_5'])
    df['a_log5cum'] = ta.above(df['log_ret_5'], df['log_retcum'])


    # df['pct_ret'] = ta.percent_return(df['close'])*100
    # df['pct_ret_5'] = ta.percent_return(df['close'], length=len)*100          # with a 5 day shift
    # df['pct_retcum'] = ta.percent_return(df['close'], cumulative=cum)*100

    # df['a_ret5cum'] = ta.above(df['pct_ret_5'], df['pct_retcum'])

    df['is_bull'] = ta.above(df['close'], df['open'])

    df.dropna(inplace=True)

    return df
```
